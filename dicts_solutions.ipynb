# Warm Up 2
Erstelle zwei gleich lange Listen:

Die erste Liste enthält Ländernnamen
`countries = ["France", "USA", ...]`

Die zweite die zugehörigen Hauptstädte:
`capitals = ["Paris", "Washington", ...]`

Der Index für das Land und seine Hauptstadt müssen gleich sein.

`countries[0] --> 'France'`

`capitals[0] --> 'Paris'`

Schreibe eine Function mit der für ein Land die Hauptstadt zurückgegeben wird.
Später sehen wir, wie dies mit *Dictionaries* einfacher gelöst wird.

````python
countries = ["France", "USA"]
capitals = ["Paris", "Washington"]
````

````python
def get_capital(country, countries, capitals):
    if country in countries: # check: ist country in der Liste?
        idx = countries.index(country) # ja: hole Index für country in der Liste
        print(f"country {country} has index {idx}")
        return capitals[idx] # gibt capital mit dem gleichen Index zurück
    else:
        print(f"country {country} is not in the list: {countries}")
        return None
````

````python
assert get_capital("France", countries, capitals) == "Paris", "The capital of France should be Paris"
assert get_capital("USA", countries, capitals) == "Washington", "The capital of USA should be Washington"
assert get_capital("Switzerland", countries, capitals) == None, "The capital of Switzerland is not in the list"
````

## Dicts Übungen
Erstelle einen `dict` für die folgenden Länder und Hauptstädte:

| *key* | *value* |
|:--|:--|
| United Kingdom | London |
| France | Paris |
| USA | Washington |
| Japan | Tokyo |

````python
# TODO: initialisiere den dictionary capitals
capitals = {
    #key:             value
    "United Kingdom": "London",
    "France":         "Paris",
    "USA":            "Washington",
    "Japan":          "Tokyo"   
}
````

````python
# Tests:
assert len(capitals) == 4, f"capitals should contain 4 entries, but there are only {len(capitals)}"
assert capitals["United Kingdom"] == "London"
````

````python
# TODO: Welche Länder sind in capitals? Tipp: Keys eines Dicts können mit <a_dict>.keys() abgefragt werden
keys = capitals.keys()
print(f"In capitals are the following countries as keys: {keys}")
````

````python
# TODO: Welche Hauptstädte sind in capitals? Tipp: Values eines Dicts können mit <a_dict>.values() abgefragt werden
values = capitals.values()
print(f"In capitals are the following capital as values: {values}")
````

````python
# TODO: Hole mit dem []-Operator einzelne Hauptstädte aus capitals
# Was passiert wenn ein ungültiger Key verwendet wird?
# Vergleiche den []-Operator mit dict.get(key)
c1 = capitals["France"]
print(f"capitals['France']: {c1}")

# Sweden ist nicht als key in capitals: dies für zu einem KeyError
try: # mit try/except kann dieser error abgefangen werden
    capitals["Sweden"] 
except KeyError:
    pass

# Anstelle capitals[<key>] kann man auch capitals.get(<key>) verwenden.
# Unterschied: dict.get(<key>) führt nicht zu einem Error, wenn der key im dict fehlt.
c2 = capitals.get("Sweden")
print(f"capitals.get('Sweden'): {c2}")
````

Schreibe eine Function, die alle Länder und ihre Hauptstädte mit `print()` ausgibt.

````python
def print_capitals(capitals):
    for country in capitals.keys():
        capital = capitals[country]
        print(f"{country}: {capital}")

print_capitals(capitals)
````

Dictionaries können auch verschachtelt werden und unterschiedliche Daten-Typen enthalten.
Erstelle einen Dict `country_details`, welcher als `Key` den Landsnamen hat als `Value` aber einen weiteren Dict `details`:

`country_details`:

| *key* | *value* |
|:--|:--|
| `<Land>` | `<details>` |

`details`:

| *key* | *value* |
|:--|:--|
| `currency` | `<str>` |
| `population` | `<int>` |
| `area` | `<double>` |
| `languages` | `<list>` |
| `neighbours` | `<list>` |

````python
country_details = {}
details = {}

# Iceland
details = {}
details["currency"] = "icelandic crown"
details["population"] = 364134
details["area"] = 103000.0
details["languages"] = ["icelandic"]
details["neighbours"] = []
country_details["Iceland"] = details

# Switzerland
country_details.update(Switzerland={"currency": "CHF",
        "population": 8570000,
        "area": 41285.0,
        "languages": ["german", "french", "italian", "romansh"],
        "neighbours": ["Austria", "France", "Germany", "Italy", "Lichtenstein"]
                                   })
````

Schreibe eine Funktion `print_details` welche pro Land alle Details ausgibt

````python
def print_details(country_details):
    for country_name in country_details.keys():
        print(f"{country_name}")
        print(f"---------------------------------------------")
        country = country_details[country_name]
        for detail in country.keys():
            print(f"{detail}\t{country[detail]}")
        print("\n")

print_details(country_details)
````
